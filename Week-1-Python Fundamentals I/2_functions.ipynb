```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def calculate_bonus(self):
        return self.salary*0.10
emp1 = Employee("Mohan", 90909)
emp1.calculate_bonus()
```

Class is a single entity , a blue print 
emp1 is an object or instance of the class Employee

```python
from datetime import date
class Human:
    def __init__(self, name, birth_year):
        self.name=name
        self.birth_year = birth_year
    def get_age(self):
        current_year = date.today().year
        return current_year-self.birth_year
    def introduce(self):
        print(f"Hi, any name is {self.name}. I am {self.get_age()} years old.")
velak = Human("Dhaval",2000)
velak.get_age()
```
